```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn import model_selection
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV
from matplotlib import rc
from matplotlib import cm
from matplotlib import rc
from mpl_toolkits.axes_grid1.inset_locator import inset_axes  
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)  
import shap
xscaler = MinMaxScaler()
```

```python
file_path = '../input_data/full_set.csv'
model_output = 'fwhm'
# import data frame for pre-processing
df = pd.read_csv(file_path)
df.head()
```

# Data pre-processing

```python
%%time
# assign input variables and target variable, eliminates static values from the data set
inputs = df.loc[:, ('Solvent Ratio', 'Amount of additive (mg/ml)')]
target = df['FWHM']

# Initializes label encoder
le_solventRatio = LabelEncoder()

# encode non-numeric type inputs
inputs['Solvent Ratio_n'] = le_solventRatio.fit_transform(inputs['Solvent Ratio'])
# remove non-encoded data from dataset 
inputs_n = inputs.drop(['Solvent Ratio'],axis='columns')

mapping = {'100': 1, '90/10': 0.9, '80/20': 0.8, '70/30': 0.7, '60/40': 0.6, '50/50': 0.5}
inputs['Solvent Ratio'].replace(mapping, inplace=True)
inputs = inputs.drop(['Solvent Ratio_n'],axis='columns')
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(inputs,target,
                                                    test_size=0.2,random_state=10)
```

```python
import numpy as np

def trim_array(array, lower_bound, upper_bound):
    """
    Trims the array to include only elements between lower_bound and upper_bound.
    
    Parameters:
    array (np.ndarray): The input NumPy array.
    lower_bound (float): The lower bound for trimming.
    upper_bound (float): The upper bound for trimming.
    
    Returns:
    np.ndarray: A trimmed array with elements between lower_bound and upper_bound.
    """
    # Boolean indexing to filter elements within the specified bounds
    trimmed_array = array[(array >= lower_bound) & (array <= upper_bound)]
    return trimmed_array


# Define the lower and upper bounds
lower_bound = 0.002
upper_bound = 0.003

# Trim the array
trim_y_test = trim_array(target, lower_bound, upper_bound)
trim_rfrfull = trim_array(rfr_full, lower_bound, upper_bound)
```
