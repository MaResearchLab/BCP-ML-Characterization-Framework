# Import packages and load data

```python
import numpy as np
import pandas as pd
from pydataset import data
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import model_selection
from statsmodels.tools.eval_measures import mse
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df = pd.read_csv('data.csv')
df.head()
```

# Data pre-processing

```python
inputs = df.drop(['pho_intens'],axis='columns')
target = df['pho_intens']
```

```python
X = inputs.to_numpy()
y = target.to_numpy()
```

# Model implementation

```python
#Grid searcher
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
svr_kf = KFold(n_splits=7, random_state=42, shuffle=True)
```

```python
data_splits_object = svr_kf.split(X, y)
```

```python
my_split_data=next(data_splits_object)
print("Train indicies are: ", my_split_data[0][0:10])
print("Test indicies are: ", my_split_data[1][0:10])
```

```python
y_pred_train = []
y_pred_test = []
y_expec_train = []
y_expec_test = []
```

```python
from sklearn.metrics import r2_score
scaler = MinMaxScaler()

for train_index, test_index in svr_kf.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    scaler.fit(X_train)
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)
    
    svr = svm.SVR(kernel='rbf', C=1000, epsilon=10, gamma='scale')
    svr.fit(X_train, y_train)
    
    trained_this_fold = svr.predict(X_train)
    predicted_this_fold = svr.predict(X_test)
    accuracy = r2_score(y_test, predicted_this_fold)
    print("R2 from this fold is: " + accuracy.__str__())
    
    # Store the values from each fold
    y_pred_train.extend(trained_this_fold)
    y_pred_test.extend(predicted_this_fold)
    y_expec_test.extend(y_test)
    y_expec_train.extend(y_train)
    
accuracy = r2_score(y_expec_test, y_pred_test)
print("R2 from all folds is: " + accuracy.__str__())
```

```python
# Convert data lists to arrays
y_pred_train = np.array(y_pred_train)
y_pred_test = np.array(y_pred_test)
y_expec_train = np.array(y_expec_train)
y_expec_test = np.array(y_expec_test)
```

# Visualization

```python
import matplotlib.pyplot as plt
from matplotlib import rc

actual_x = np.arange(1, 513)

# Create a scatter plot
plt.scatter(actual_x, y_pred_test, color='blue', label='Actual')

# Add a diagonal line for reference (perfect prediction)
plt.plot(actual_x, y_pred_test, linestyle='--', color='red', label='Prediction')

# Add labels and a legend
plt.xlabel('Actual Data', fontsize='20')
plt.ylabel('Predicted Data', fontsize='20')
rc('xtick', labelsize=15)   
rc('ytick', labelsize=15)
plt.legend(fontsize='12.5')

# Show the plot
plt.title('Actual vs. Predicted Data')
plt.grid(False)
plt.show()
```

```python
# Create a scatter plot
plt.scatter(y_expec_train, y_pred_train, color='blue', label='Predicted Train', alpha=0.7)
plt.scatter(y_expec_test, y_pred_test, color='green', label='Predicted Test', alpha=0.7)

# Add a diagonal line for reference (perfect prediction)
plt.plot([min(y_expec_test), max(y_expec_test)], [min(y_expec_test), max(y_expec_test)], linestyle='--', color='red', label='Perfect Prediction')

# Add labels and a legend
plt.xlabel('Actual Data', fontsize='20')
plt.ylabel('Predicted Data', fontsize='20')
rc('xtick', labelsize=15)   
rc('ytick', labelsize=15)
plt.legend(fontsize='12.5')

# Show the plot
plt.title('Actual vs. Predicted Data')
plt.grid(False)
plt.show()
```

```python
# Residuals Plot!!!

res_train = y_expec_train - y_pred_train
res_test = y_expec_test - y_pred_test

# Create a residuals plot
plt.figure(figsize=(8, 6))
plt.scatter(y_pred_train, res_train, color='blue', alpha=0.7, label='Training Residuals')
plt.scatter(y_pred_test, res_test, color='green', alpha=0.7, label='Testing Residuals')
plt.axhline(y=0, color='red', linestyle='--', linewidth=1.5)
plt.xlabel("Predicted Values", fontsize='20')
plt.ylabel("Residuals", fontsize='20')
rc('xtick', labelsize=15)   
rc('ytick', labelsize=15)
plt.legend(fontsize='12.5')
plt.grid(False)
plt.show()
```
