```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from matplotlib import rc
import textwrap

%matplotlib inline
rc('text', usetex=False)
rc('mathtext', fontset='cm')
rc('xtick', labelsize=20)   
rc('xtick.major', size=7)  
rc('xtick.minor', size=4)
rc('xtick.major', width=2)
rc('xtick.minor', width=2)
rc('ytick.major', width=2)
rc('ytick.minor', width=2)
rc('ytick', labelsize=20)
rc('ytick.major', size=7)
rc('ytick.minor', size=4)
rc('axes', labelsize=25) 
rc('axes', linewidth=2) 
rc('font',family='sans serif')
rc('font', style='normal')
rc('font', weight='500')
rc('font', size='15')
rc('axes', labelweight='500')
rc('axes.spines', **{'right':True, 'top':True}) 
plt.rcParams['font.family'] = 'Arial'
# Example data: R² scores for each model type for four outputs
# Assuming the structure is a dictionary where keys are output names
# and values are lists of R² scores for the corresponding models
r2_scores = {
    'Linear': [0.69, 0.19, 0.49, 0.09],
    'Lasso': [0.70, 0.03, 0.50, 0.05],
    'Polynomial': [0.81, 0.35, 0.55, 0.09],
    'Power': [0.79, 0.28, 0.54, 0.14]
}

# Model types
model_types = ['GISAXS Domain Spacing', 'FWHM', 'AFM Domain Spacing', 'Grain Size']

# Number of outputs
n_outputs = len(r2_scores)

# Bar width
bar_width = 0.2

# Positions of the bars on the x-axis
index = np.arange(len(model_types))

# Creating subplots
fig, ax = plt.subplots()

# Colors for each group
colors = ['blueviolet', 'mediumvioletred', 'violet', 'lavenderblush']
# Plotting each group
for i, (output, scores) in enumerate(r2_scores.items()):
    ax.bar(index + i * bar_width, scores, bar_width, label=output, color=colors[i], alpha=0.7, 
                      edgecolor='grey')
# Wrapping the x-tick labels
wrapped_labels = [textwrap.fill(label, 8) for label in model_types]
# Adding labels, title, and legend
# ax.set_xlabel('')
ax.set_ylabel('R² Score')
#ax.set_title('R² Scores by Model Type and Output')
ax.set_xticks(index + bar_width * (n_outputs - 1) / 2)
ax.set_xticklabels(wrapped_labels)
ax.legend(frameon=False)

# Display the plot
plt.tight_layout()
plt.savefig('quant_comp.png', dpi=500, format='png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from matplotlib import rc
import textwrap

r2_scores = {
    'Random Forest': [0.84, 0.67, 0.55, 0.16],
    'Support Vector': [0.70, 0, 0.50, 0.19],
    'XGBoost': [0.83, 0.58, 0.49, 0.16],
    'Multilayer Perceptron': [0.81, 0.21, 0.56, 0.09]
}

# Model types
model_types = ['GISAXS Domain Spacing', 'FWHM', 'AFM Domain Spacing', 'Grain Size']

# Number of outputs
n_outputs = len(r2_scores)

# Bar width
bar_width = 0.2

# Positions of the bars on the x-axis
index = np.arange(len(model_types))

# Creating subplots
fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
# Colors for each group
colors = ['blueviolet', 'mediumvioletred', 'violet', 'lavenderblush']
# Plotting each group
for i, (output, scores) in enumerate(r2_scores.items()):
    ax.bar(index + i * bar_width, scores, bar_width, label=output, color=colors[i], alpha=0.7, 
                      edgecolor='grey')
# Wrapping the x-tick labels
wrapped_labels = [textwrap.fill(label, 8) for label in model_types]
# Adding labels, title, and legend
# ax.set_xlabel('')
ax.set_ylabel('R² Score')
#ax.set_title('R² Scores by Model Type and Output')
ax.set_xticks(index + bar_width * (n_outputs - 1) / 2)
ax.set_xticklabels(wrapped_labels)
ax.legend(frameon=False)

# Display the plot
plt.tight_layout()
plt.savefig('ml_comp.png', dpi=500, format='png')
plt.show()
```
