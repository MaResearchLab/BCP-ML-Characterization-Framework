# Import packages and load data

```python
import pandas as pd
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn import model_selection
from statsmodels.tools.eval_measures import mse
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
xscaler = MinMaxScaler()
yscaler = MinMaxScaler()
sns.set()
```

```python
df = pd.read_csv('UpdatedOptimizedParameters.csv')
df.head()
```

# Data pre-processing

```python
inputs = df.drop(['Ordered/Disorded','Peak Position','FWHM','Sample Name','Polymer Component',
                  'domain', 'Polymer Conc.(mg/ml)','Additive type', 'Solvent type'],axis='columns')
inputs.head()
```

```python
# Initializes label encoder for 2 different variables
le_solventType = LabelEncoder()
le_solventRatio = LabelEncoder()

# encode non-numeric type inputs
inputs['Amount of additive (mg/ml)_n'] = le_solventType.fit_transform(inputs['Amount of additive (mg/ml)'])
inputs['Solvent Ratio_n'] = le_solventRatio.fit_transform(inputs['Solvent Ratio'])

# remove non-encoded data from dataset 
inputs_n = inputs.drop(['Amount of additive (mg/ml)','Solvent Ratio'],axis='columns')
inputs_n.head()
```

```python
target_fw = df['FWHM']
target_dom = df['domain']
```

```python
# For Support Vector Model
X_train, X_test, y_train, y_test = train_test_split(inputs_n,target_dom,test_size=0.2,random_state=10)
```

```python
xscaler.fit(X_train)
X_train = xscaler.transform(X_train)
X_test = xscaler.transform(X_test)
```

# Build a pipeline of training

```python
from sklearn.pipeline import Pipeline
from xgboost import XGBRegressor


estimators = [
    #('encoder', TargetEncoder()),
    ('rgr', XGBRegressor(random_state=10))
]
pipe = Pipeline(steps=estimators)
pipe
```

# Set up hyperparameter tuning

```python
from skopt import BayesSearchCV
from skopt.space import Real, Categorical, Integer
import sklearn
from sklearn.model_selection import GridSearchCV

search_space = {
    'rgr__max_depth': [2, 3, 4, 5, 6, 7, 8],
    'rgr__learning_rate': [0.001, 0.01, 0.1, 1.0],
    'rgr__subsample': [0.5, 1.0],
    'rgr__colsample_bytree': [0.5, 1.0],
    'rgr__colsample_bylevel': [0.5, 1.0],
    'rgr__colsample_bynode': [0.5, 1.0],
    'rgr__reg_alpha': [0.0, 2.5, 5.0, 7.5, 10.0],
    'rgr__reg_lambda': [0.0, 2.5, 5.0, 7.5, 10.0],
    'rgr__gamma': [0.0, 2.5, 5.0, 7.5, 10.0]
}

opt = GridSearchCV(pipe, search_space, cv=3, scoring='r2')
```

# Train the model

```python
opt.fit(X_train, y_train)
```

# Evaluate the mode

```python
print(opt.best_params_)
print(opt.best_score_)

best = opt.best_estimator_
```

```python
from sklearn.metrics import r2_score

xgeval = best.predict(X_test)
print(r2_score(y_test, xgeval))
xgeval_train = best.predict(X_train)                                              
```

# Measure feature importance

```python
opt.best_estimator_.steps
```

```python
from xgboost import plot_importance

xgboost_step = opt.best_estimator_.steps[0]
xgboost_model = xgboost_step[1]
plot_importance(xgboost_model._Booster)
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc

# Create a scatter plot
plt.scatter(y_train, xgeval_train, color='blue', label='Predicted Train', alpha=0.7)
plt.scatter(y_test, xgeval, color='green', label='Predicted Test', alpha=0.7)

# Add a diagonal line for reference (perfect prediction)
plt.plot([min(y_train), max(y_test)], [min(y_train), max(y_test)], linestyle='--', color='red', label='Perfect Prediction')

# Add labels and a legend
plt.xlabel('Actual Data', fontsize='20')
plt.ylabel('Predicted Data', fontsize='20')
rc('xtick', labelsize=15)   
rc('ytick', labelsize=15)
plt.legend(fontsize='12.5')

# Show the plot
#plt.title('Actual vs. Predicted Data')
plt.grid(False)
plt.tight_layout()
plt.savefig('xgb_avp.png', format='png')
plt.show()
```
