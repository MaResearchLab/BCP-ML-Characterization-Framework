```python
import tensorflow as tf
from tensorflow.keras import layers, models

def DeeperCNN(input_shape=(128, 128, 1), num_classes=3, num_filters1=32, num_filters2=64, num_filters3=128,
              kernel_size=5, dropout_rate=0.5):
    model = models.Sequential()
    
    # Block 1
    model.add(layers.Conv2D(num_filters1, kernel_size=kernel_size, padding='same', activation='relu', input_shape=input_shape))
    model.add(layers.BatchNormalization())
    model.add(layers.Conv2D(num_filters1, kernel_size=kernel_size, padding='same', activation='relu'))
    model.add(layers.BatchNormalization())
    model.add(layers.MaxPooling2D(pool_size=(2, 2)))
    
    # Block 2
    model.add(layers.Conv2D(num_filters2, kernel_size=kernel_size, padding='same', activation='relu'))
    model.add(layers.BatchNormalization())
    model.add(layers.Conv2D(num_filters2, kernel_size=kernel_size, padding='same', activation='relu'))
    model.add(layers.BatchNormalization())
    model.add(layers.MaxPooling2D(pool_size=(2, 2)))
    
    # Block 3
    model.add(layers.Conv2D(num_filters3, kernel_size=kernel_size, padding='same', activation='relu'))
    model.add(layers.BatchNormalization())
    model.add(layers.Conv2D(num_filters3, kernel_size=kernel_size, padding='same', activation='relu'))
    model.add(layers.BatchNormalization())
    model.add(layers.MaxPooling2D(pool_size=(2, 2)))
    
    # Fully connected layers
    model.add(layers.Flatten())
    model.add(layers.Dense(512, activation='relu'))
    model.add(layers.Dropout(dropout_rate))
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dropout(dropout_rate))
    model.add(layers.Dense(num_classes, activation='softmax'))
    
    return model

# Example usage
model = DeeperCNN()
model.summary()
```

```python
import visualkeras
from PIL import ImageFont
from tensorflow.keras.layers import Conv2D, ReLU, BatchNormalization, MaxPooling2D, Flatten, Dropout, Dense
from collections import defaultdict

model = model

colormap = defaultdict(dict)
colormap[Conv2D]['fill'] = 'blueviolet'
colormap[BatchNormalization]['fill'] = 'violet'
colormap[ReLU]['fill'] = 'lavenderblush'
colormap[MaxPooling2D]['fill'] = 'darkblue'
colormap[Flatten]['fill'] = 'royalblue'
colormap[Dropout]['fill'] = 'mediumvioletred'
colormap[Dense]['fill'] = 'pink'
#visualkeras.layered_view(model).show
font = ImageFont.truetype("arial.ttf", 50)
visualkeras.layered_view(model, font=font, spacing=25, color_map=colormap, legend=True, to_file='CNN_framework_legend.png')
```
