```python
import cv2
import os
import argparse
import glob

def count_pixels(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        print(f"Error: Could not read image {image_path}")
        return None, None
    
    black_pixels = (image == 0).sum()
    white_pixels = (image == 255).sum()
    
    return black_pixels, white_pixels

def process_images(image, image_name):
    black, white = count_pixels(image)
    # Save the number of pixels in each group to a text document
    with open('line_area.txt', "a") as file:
        if black is not None and white is not None:
            file.write(f"Sample {image_name} phase ratio: {(black+white) / black}\n")

input_folder = 'data_set/'

# List of desired extensions
extensions = ["*.tiff", "*.png"]

# Find all files with matching extensions
image_files = [f for ext in extensions for f in glob.glob(os.path.join(input_folder, ext))]

#Create a loop for all samples
for image_file in image_files:

    image_name = os.path.splitext(os.path.basename(image_file))[0]
    print(f"Processing {image_name}")
    
    process_images(image_file, image_name)
```
