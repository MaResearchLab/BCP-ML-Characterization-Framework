```python
%%time
import numpy as np
import warnings
warnings.filterwarnings("ignore") # stop printing a precision warning

import math
from scipy.stats import mode
import os
import csv
import matplotlib.pylab as pylab
import matplotlib.pyplot as plt
import pdb

params = {'legend.fontsize': 'x-large',
		 'axes.labelsize': 'x-large',
		 'axes.titlesize':'x-large',
		 'xtick.labelsize':'x-large',
		 'ytick.labelsize':'x-large'}
pylab.rcParams.update(params)

import scipy

import skimage
from skimage import color
import skimage.util
from skimage.transform import resize
from sklearn.cluster import DBSCAN


from PIL import Image 
from voronoi_v7 import analyze_image


# Walk through the AFM image directory and find all images
#jpgs = []
pngs =[]
image_names = []
image_sizes = []

subimages = False # Perform the analysis on 4 subimages for every image, in addition to full-image analysis

# where the images are stored
data_file_location = '../labelme/binned/dots/'
# each image will get its own subfolder located here
results_file_location = "voronoi_outputs/"
#data_csv_name = "image_analysis_output"
data_csv_name = "image_analysis_output_knn_spam"


for root, dirs, files in os.walk(data_file_location):
    for file in files:
        if 'D_AC_bin.png' in file:
            im = Image.open(root + '/' + file)
            ################################ added by me ############################
            im=im.convert("RGB") # convert the image to RGB (drops alpha channel/transparency if present)
            data = 1 - color.rgb2gray(skimage.img_as_float(im))
            pngs.append(data)
            image_names.append(file)
            file = file.replace('.','_') # use either a dash or an underscore to separate items
            name_split = file.split('_')
##########################  added because the code won't run without knowing the image size
            image_sizes.append(1)
            im.close() #avoid a "too many files open" error

for i in range(len(pngs)):

    image_data = pngs[i]
    name = image_names[i]
    size = image_sizes[i]
    print("---------------------------------------------")
    print(f"          processing image {name}")
    print("---------------------------------------------")

# this is the analysis for the full image
    results_dict = analyze_image(image_data, name, size, True, False, results_file_location, threshold_edge=0.025) 

    with open(r''+ data_csv_name +'.csv', 'a', newline='') as csvfile: 
	
        fieldnames = ['name','dat_pmer', 'dat_pmer_index','thy_pmer','thy_pmer_index', 'pxls_to_um','img_size','periodicity',
						'grain_size', 'mean_morph','block_ratio',
						'percent_lines_dark','junct_density_dark','end_density_dark',
						'percent_lines_light', 'junct_density_light','end_density_light']

        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writerow(results_dict)
	# use periodicity from the full image to crop into subimages

    if subimages == True:
        width, height = image_data.shape
        n_subimages = 2
        pxls_to_um = results_dict['pxls_to_um']
        if results_dict['periodicity'] != 'f':
            period = results_dict['periodicity']*pxls_to_um/1000
            periods_per_side = 20
            for n in range(n_subimages): # just stick with 4 corners for all images
                if n == 0:
                    n_low = 0
                    n_high = int(0 + period*periods_per_side)
                else:
                    n_high = widthn_low = int(width - period*periods_per_side)
                for m in range(n_subimages):
                    if m == 0:
                        m_low = 0
                        m_high = int(0 + period*periods_per_side)
                    else:
                        m_high = width
                        m_low = int(width - period*periods_per_side)
                    subimage_data = image_data[n_low:n_high, m_low:m_high]
                    sub_name = name+'_sub_'+str(n)+str(m)

					# find adjusted size using pxls_to_um
                    sub_width, sub_height = subimage_data.shape

					# perform image analysis
                    sub_results_dict = analyze_image(subimage_data, sub_name, sub_width/pxls_to_um, True, 
                                                     False, results_file_location, threshold_edge=0.025) # this is the analysis for the sub image
                    with open(r''+ data_csv_name +'.csv', 'a', newline='') as csvfile: 
                        fieldnames = ['name','dat_pmer', 'dat_pmer_index','thy_pmer','thy_pmer_index', 'pxls_to_um',
                        'img_size','periodicity',
						'grain_size', 'mean_morph','block_ratio',
						'percent_lines_dark','junct_density_dark','end_density_dark',
						'percent_lines_light', 'junct_density_light','end_density_light']
                    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
                    writer.writerow(sub_results_dict)
```
