```python
%%time
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Import packages
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
from scipy.signal import find_peaks
import matplotlib.pyplot as plt
from openpyxl import Workbook, load_workbook

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Create a sheet for optimized parameters
batch = 'Batch1AB'
workbook = load_workbook(f"./raw_data/{batch}.xlsx")
data_sheet = workbook.active
sheet_op = workbook.create_sheet("Optimized Parameters_new", 0)
sheet_op['A1'].value = "Sample Name"
sheet_op['B1'].value = "Toluene:THF Ratio"
sheet_op['C1'].value = "Polymer Concentration"
sheet_op['D1'].value = "Additive Type"
sheet_op['E1'].value = "Additive:Polymer Ratio"
sheet_op['F1'].value = "Ordered/Disorded"
sheet_op['G1'].value = "Peak Position"
sheet_op['H1'].value = "FWHM"
sheet_op['I1'].value = "R2"

def user_equation(x, a, b, c, d, e, f):
                # Modify this equation according to your needs
                return a + b*np.power(x, -c) + d*np.exp((-4*np.log(2)*(x-e)**2)/((f)**2))

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Fill array with column titles
# Get the maximum column index
max_column = data_sheet.max_column
# Initialize a list to store the text of each cell in the first row
first_row_text = []
# Iterate through each cell in the first row
for column in data_sheet.iter_cols(min_row=1, max_row=1, max_col=max_column):
    for cell in column:
        # Append the text of the cell to the list
        first_row_text.append(cell.value)
        
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Determines ranges of each data set
data_ranges = []
start_col = 0
cur_col = 0
for column in data_sheet.iter_cols():
    column_values = [cell.value for cell in column]
    if column_values.count(None) >= len(column_values)-1:
        if cur_col - start_col > 2:
            data_ranges.append((start_col, cur_col - 1))
        start_col = cur_col + 1            
    cur_col += 1

if start_col != cur_col:
    data_ranges.append((start_col, cur_col - 1))
data_num = 0
for start_col, end_col in data_ranges:
    for i in range(end_col-start_col):
        data_num += 1
# print(data_ranges)
        
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Find intial guess values for FWHM and peak position
data_num = 0
for start_col, end_col in data_ranges:
    for i in range(end_col-start_col):
        data_num += 1    
        #Find intial guess values for FWHM and peak position
        df = pd.read_excel(f"./raw_data/{batch}.xlsx")
        x_index = first_row_text[start_col] 
        y_index = first_row_text[i+1+start_col]
        print(f"Sample name: {y_index}")
        # Scale the axes
        x_values = df[x_index]*1000 #Record the x data and scale
        y_values = df[y_index]*10
        
# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Find the index of the maximum peak
        # Find the indices of the peaks
        max_y = None  # Variable to store the maximum 'y' value
        max_x = None  # Variable to store the corresponding 'x' value
        # Iterate over the 'y' values and check the condition
        for i in range(1, len(y_values) - 1):
            if y_values[i] > y_values[i-1] and y_values[i] > y_values[i+1]:
                if max_y is None or y_values[i] > max_y:
                    if 21 > x_values[i] > 16:
                        max_y = y_values[i]
                        max_x = x_values[i]
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Check if any peaks are found
        # print(f"Peak Guess: {(max_x/1000):.4}")
        print(f"Peak Guess: {(max_x)}")
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Trim data set around the peak            
        if max_x != None:    
            x_data = []
            y_data = []
            for x, y in zip(x_values, y_values):
                if max_x-4 < x < max_x+4:
                #if 10 < x < 30:
                    x_data.append(x)
                    y_data.append(y)
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Define the data fitting equation
            # def user_equation(x, a, b, c, d, e, f):
            #     # Modify this equation according to your needs
            #     return a + b*np.power(x, -c) + d*np.exp((-4*np.log(2)*(x-e)**2)/((f)**2))
            try:
                # Perform curve fitting
                popt, pcov = curve_fit(user_equation, x_data, y_data, p0 = [0, 0, 0, max_y, max_x, 1], maxfev =25000)
    
                # Extract the optimized parameters
                a_opt, b_opt, c_opt, d_opt, e_opt, f_opt = popt
    
                # Calculate the fitted y values
                y_fit = user_equation(x_data, a_opt, b_opt, c_opt, d_opt, e_opt, f_opt)
    
                # Calculate the R-squared value
                residuals = y_data - y_fit
                ss_residuals = np.sum(residuals ** 2)
                ss_total = np.sum((y_data - np.mean(y_data)) ** 2)
                r_squared = 1 - (ss_residuals / ss_total)
                
                x_data[:] = [x / 1000 for x in x_data]
                y_data[:] = [x * 100 for x in y_data]
                y_fit[:] = [x * 100  for x in y_fit]
                print(f"Peak Position: {(e_opt/1000):.4}")
                print(f"FWHM: {abs(f_opt/1000):.4}")
                print(f"R-Squared: {r_squared:.4}")
                # Plot the scatter plot of original data points and the fitting line
                plt.scatter(x_data, y_data, label='Original Data')
                plt.plot(x_data, y_fit, color='red', label='Fitting Line')
                plt.xlabel('Q vector [$A^{-1}$]')
                plt.ylabel('Intensity / a.u.')
                plt.legend()
                plt.show()
    
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Write optimized parameters to sheet created above
                #cell_name = f"A{i+1}"
                sheet_op[f"A{data_num+1}"].value = y_index
                # sheet_op[f"F{data_num+1}"].value = "Ordered"
                sheet_op[f"G{data_num+1}"].value = e_opt/1000
                sheet_op[f"H{data_num+1}"].value = abs(f_opt/1000)
                sheet_op[f"I{data_num+1}"].value = r_squared
            except RuntimeError:
                print(f"Sample {y_index}: Maximum number of function evaluations exceeded. Skipping this sample.")
                #cell_name = f"A{i+1}"
                sheet_op[f"A{data_num+1}"].value = y_index
                # sheet_op[f"F{data_num+1}"].value = "Ordered"
                sheet_op[f"G{data_num+1}"].value = 'N/A'
                sheet_op[f"H{data_num+1}"].value = 'N/A'
                sheet_op[f"I{data_num+1}"].value = 'N/A'
        else:
            sheet_op[f"A{data_num+1}"].value = y_index
            # sheet_op[f"F{data_num+1}"].value = "Disordered"
            sheet_op[f"G{data_num+1}"].value = "N/A"
            sheet_op[f"H{data_num+1}"].value = "N/A"
            sheet_op[f"I{data_num+1}"].value = 'N/A'

#workbook.save('Scatter Parameters_new.xlsx')
```
